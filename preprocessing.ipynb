```python
import pandas as pd
```

```python
df=pd.read_csv('trek_data.csv')
df.head()
```

```python
print(df.isnull().sum())
```

```python
print("Available columns:", df.columns.tolist())
```

```python
df['Est_time'] = df['Est_time'].fillna(df['Est_time']).mean()
```

```python

df = df.drop(['link_AllTrails','image','description'], axis=1)
df
```

```python
df['number_of_reviews'] = df['number_of_reviews'].abs()
df
```

```python
df['Length'] = df['Length'].str.extract(r'(\d+\.?\d*)').astype(float)
df
```

```python
df.rename(columns={'Length': 'Length (in km)'}, inplace=True)
```

```python
df
```

```python
df[['City', 'State', 'Country']] = df['Location'].str.split(',', expand=True)
df['City'] = df['City'].str.strip()
df['State'] = df['State'].str.strip()
df['Country'] = df['Country'].str.strip()

df = df.drop('Location', axis=1)
df
season_mapping = {
    'Himachal Pradesh': 'April - June, September - November',
    'Uttarakhand': 'March - June, September - November',
    'Maharashtra': 'October - February',
    'Karnataka': 'October - February',
    'Kerala': 'September - March',
    'Jammu and Kashmir': 'May - October',
    'West Bengal': 'October - March',
    'Tamil Nadu': 'November - February',
    'Goa': 'November - February',
   
}

# Add Best_Season column based on State
df['Best_Season'] = df['State'].map(season_mapping).fillna('All Year')

```

```python
df
```

```python

# Basic cleaning
df['Tags'] = df['Tags'].fillna('').str.lower()
df['Tags'] = df['Tags'].str.replace(r"[\[\]']", '', regex=True)

# Split to lists
df['tags_list'] = df['Tags'].str.split(',')

# Strip spaces
df['tags_list'] = df['tags_list'].apply(lambda x: [tag.strip() for tag in x] if isinstance(x, list) else [])

# Remove duplicates
df['tags_list'] = df['tags_list'].apply(lambda x: list(set(x)))
df
```

```python
# empty values are treated as NaN
df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')
df['longitude'] = pd.to_numeric(df['longitude'], errors='coerce')
```

```python
# Drop rows where lat/lon are missing
df = df[df['latitude'].notnull() & df['longitude'].notnull()]
```

```python
# Filling missing values
df['current_temperature'].fillna(df['current_temperature'].median(), inplace=True)
df['current_temperature']
```

```python
df['current_windspeed'].fillna(df['current_windspeed'].median(), inplace=True)
df['current_windspeed']
```

```python
df['current_winddirection'].fillna(df['current_winddirection'].mean(), inplace=True)
df['current_winddirection']
```

```python

df['current_weather_code'].fillna(df['current_weather_code'].mode()[0], inplace=True)
df['current_weather_code']
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()

# Fit and transform your list column
tag_encoded = mlb.fit_transform(df['tags_list'])

# Create DataFrame
tag_df = pd.DataFrame(tag_encoded, columns=mlb.classes_)

# Concatenate with original
df_encoded = pd.concat([df, tag_df], axis=1)

```

```python
df_encoded.head()
```

```python
# Create a working copy
df_scaled = df_encoded.copy()

# IQR outlier removal on trail length
Q1, Q3 = df_scaled['Length (in km)'].quantile([0.25, 0.75])
IQR = Q3 - Q1
low = Q1 - 1.5 * IQR
high = Q3 + 1.5 * IQR

# Filter out outliers
df_scaled = df_scaled[df_scaled['Length (in km)'].between(low, high)].reset_index(drop=True)

print("Remaining rows:", len(df_scaled))
```

```python
def remove_outliers_iqr(df, columns):

    df_filtered = df.copy()
    for col in columns:
        Q1 = df_filtered[col].quantile(0.25)
        Q3 = df_filtered[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df_filtered = df_filtered[df_filtered[col].between(lower, upper)]
    return df_filtered.reset_index(drop=True)
```

```python
outlier_cols = ['Length (in km)', 'number_of_reviews', 'current_temperature', 'current_windspeed']
df_clean = remove_outliers_iqr(df_encoded, outlier_cols)
```

```python
def classify_weather(code):
    if code in [1, 2]: return 'Clear'
    elif code in [3]: return 'Cloudy'
    elif 45 <= code <= 82: return 'Rainy/Snowy'
    else: return 'Other'

df_clean['weather_type'] = df_clean['current_weather_code'].apply(classify_weather)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numeric_cols = df_encoded.select_dtypes(include=['number']).columns


df_encoded[numeric_cols] = scaler.fit_transform(df_encoded[numeric_cols])
df_encoded.head()
```
